```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./a/CPU_V.csv", index_col=False, names=["cpu"])
df["times"] = pd.read_csv("./a/T_V.csv", names=["times"])

df.set_index("times")
```

```python
plt.style.use("seaborn")
plot = sns.lineplot(x="times", y="cpu", data=df, marker='o')
plot.set(yscale="log")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

t = np.asarray(df["times"], dtype=float).reshape(-1, 1)
cpu = np.log(df["cpu"])

lr.fit(t, cpu)
x = np.linspace(50, 200, 20)
y = np.exp(lr.intercept_ + lr.coef_*x)
plt.style.use("seaborn")
sns.lineplot(x="times", y="cpu", data=df, marker='o')
sns.lineplot(x=x, y=y, linestyle='--', alpha=0.8)
plt.yscale("log")
plt.xlabel("T")
plt.ylabel("CPU seconds")
plt.title("CPU seconds as a function of T with relaxed integrality requirements on z")
plt.show()
```
